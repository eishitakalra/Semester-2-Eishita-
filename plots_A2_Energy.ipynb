```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
df = pd.read_csv("plot1.csv")  # Replace with actual filename

# Define the x-axis (time periods)
x = df["t"]

# Define the different power sources
ccgt = df["Power Output of CCGT"]
coal = df["Power Output of CCGT+Coal"] - df["Power Output of CCGT"]
ocgt = df["Power Output of CCGT+Coal+OCGT"] - df["Power Output of CCGT+Coal"]
wind = df["Power Output of CCGT+Coal+OCGT+Wind"] - df["Power Output of CCGT+Coal+OCGT"]
solar = df["Power Output of CCGT+Coal+OCGT+Wind+Solar"] - df["Power Output of CCGT+Coal+OCGT+Wind"]
demand = df["Demand"]

# Set high resolution and figure size
plt.figure(figsize=(12, 6), dpi=300)

# Stack the generation sources with anti-aliasing
plt.fill_between(x, 0, ccgt, label="CCGT", color="royalblue", alpha=0.7, antialiased=True)
plt.fill_between(x, ccgt, ccgt + coal, label="Coal", color="gold", alpha=0.7, antialiased=True)
plt.fill_between(x, ccgt + coal, ccgt + coal + ocgt, label="OCGT", color="orange", alpha=0.7, antialiased=True)
plt.fill_between(x, ccgt + coal + ocgt, ccgt + coal + ocgt + wind, label="Wind", color="seagreen", alpha=0.7, antialiased=True)
plt.fill_between(x, ccgt + coal + ocgt + wind, ccgt + coal + ocgt + wind + solar, label="Solar", color="orchid", alpha=0.7, antialiased=True)

# Plot demand as a line with high quality
plt.plot(x, demand, label="Demand", color="black", linewidth=2, linestyle="-", antialiased=True)

# Labels and title
plt.xlabel("Time Period")
plt.ylabel("Power Output (MW)")
plt.title("Power Generation ")
plt.legend()

# Improve font and layout for crisp text
plt.rcParams["font.size"] = 14
plt.rcParams["font.family"] = "DejaVu Sans"
plt.tight_layout()
plt.grid(True, linestyle="--", alpha=0.5)

# Save as a high-resolution vector file (sharp like Gnuplot)
# plt.savefig("Power_Output_Stacked.pdf", format='pdf', bbox_inches='tight')
# plt.savefig("Power_Output_Stacked.svg", format='svg', bbox_inches='tight')

# Show plot
plt.show()
# "darkred", "lightcoral", "orangered", "tomato", "salmon"
# "darkgreen", "limegreen", "forestgreen", "seagreen", "palegreen"
# "navy", "royalblue", "skyblue", "deepskyblue", "dodgerblue"
# "gold", "orange", "darkorange", "chocolate", "sienna"
# "purple", "indigo", "violet", "plum", "orchid"
# "brown", "tan", "peru", "beige", 
```

```python
import matplotlib.pyplot as plt

# List of extended color names
colors = [
    "darkred", "lightcoral", "orangered", "tomato", "salmon",
    "darkgreen", "limegreen", "forestgreen", "seagreen", "palegreen",
    "navy", "royalblue", "skyblue", "deepskyblue", "dodgerblue",
    "gold", "orange", "darkorange", "chocolate", "sienna",
    "purple", "indigo", "violet", "plum", "orchid",
    "brown", "tan", "peru", "beige", "wheat"
]

# Create a plot with squares for each color
fig, ax = plt.subplots(figsize=(10, 10))
ax.set_xticks([])  # Hide x ticks
ax.set_yticks([])  # Hide y ticks

# Plot each square with a different color
n = len(colors)
rows = 5  # Number of rows
cols = 6  # Number of columns

for i, color in enumerate(colors):
    row = i // cols
    col = i % cols
    ax.add_patch(plt.Rectangle((col, rows - row - 1), 1, 1, color=color))

plt.xlim(0, cols)
plt.ylim(0, rows)

# Show the plot
plt.show()
```
